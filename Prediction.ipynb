```
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.cm as cm
import matplotlib.pyplot as plt
 
# Importing data
##working data
data = pd.read_csv("test.csv")
 
data.head()
```

```
# Removing character "F" present in data
data=data[data.views!='F']
data=data[data.likes!='F']
data=data[data.dislikes!='F']
data=data[data.comment!='F']

data.head()
```

```
##copy of data to add adview later after prediction
new_data = data.copy(deep=False)
```

```
# Assigning a number to each category
category={'A': 1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7,'H':8}
data["category"]=data["category"].map(category)

data.head()
```

```
# Convert values to integers for views, likes, comments and dislikes
data["views"] = pd.to_numeric(data["views"])
data["comment"] = pd.to_numeric(data["comment"])
data["likes"] = pd.to_numeric(data["likes"])
data["dislikes"] = pd.to_numeric(data["dislikes"])

data.head()
```

```
# Endoding features like Category, Duration
from sklearn.preprocessing import LabelEncoder
data['duration']=LabelEncoder().fit_transform(data['duration'])
data['published']=LabelEncoder().fit_transform(data['published'])
```

```
# Convert Time in sec for duration
import datetime
import time
def checki(x):
    y = x[2:]
    h = ''
    m = ''
    s = ''
    mm = ''
    P = ['H','M','S']
    for i in y:
        if i not in P:
            mm+=i
        else:
            if(i=="H"):
                h = mm
                mm = ''
            elif(i == "M"):
                m = mm
                mm = ''
            else:
                s = mm
                mm = ''
    if(h==''):
        h = '00'
    if(m == ''):
        m = '00'
    if(s==''):
        s='00'
    bp = h+':'+m+':'+s
    return bp
mp = pd.read_csv("test.csv")["duration"]
time = mp.apply(checki)

def func_sec(time_string):
    h, m, s = time_string.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)

time1=time.apply(func_sec)
data["duration"]=time1


data.head()
```

```
#Dropping vidid column from dataframe
data=data.drop(["vidid"],axis=1)

data.head()
```

```
# Normalise Data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data=scaler.fit_transform(data)
```

```
#importing library
import joblib

#loading model from training
model = joblib.load("decisiontree_youtubeadview.pkl")

#predicting values
result = model.predict(data)

#attaching result array to result database
new_data["adview"]=result
new_data
```
